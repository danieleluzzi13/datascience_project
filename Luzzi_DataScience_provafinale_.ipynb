Es. 6 - 5 grafici comprensivi di titolo, labels e interpretazione (commento per ciascuno) di cui almeno 1 boxplot, 1 istogramma e 1 scatterplot usando sia Matplotlib che Seaborn.

Importo le librerie matplotlib e seaborn per la creazione di grafici

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Converto tutte le righe delle variabili '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]' da 'object' a 'float64', quindi in valori numerici, così il dataframe sarà pronto per visualizzare grafici e analisi

```python
anni = ['2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]']
for col in anni:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

Boxplot con Seaborn.
L'asse x rappresenta i diversi continenti, mentre l'asse y rappresenta l'accesso all'elettricità nel 2016. Ogni "scatola" nel grafico mostra i seguenti elementi per ciascun continente: a linea centrale rappresenta la mediana (il valore medio) dell'accesso all'elettricità per quel continente.
La scatola rappresenta il primo e il terzo quartile (il 50% centrale dei dati).
I "baffi" (whiskers) si estendono dalla scatola per mostrare l'intervallo dei dati, escludendo gli outlier. I punti individuali rappresentano gli outlier, ovvero i valori che si discostano significativamente dalla maggior parte dei dati. Se un punto è molto più in alto rispetto alla scatola e ai baffi, significa che quel paese del continente ha un accesso all'elettricità molto più alto rispetto alla maggior parte dei paesi del suo continente. Se un punto è molto più in basso rispetto alla scatola e ai baffi, significa che quel paese del continente ha un accesso all'elettricità molto più basso rispetto alla maggior parte dei paesi del suo continente. Guardando il grafico si osserva come il continente Africano sia quello più penalizzato, risultando non competitivo dal punto di visto digitale e carente nell'accesso all'energia, dunque escluso da un vita agiata e risultando anacronistico rispetto ai paesi più ricchi e innovativi del globo. Una grande disuguglianza, ma anche una grande opportunità considerando l'Africa il continente più fecondo, in quanto in aumento demografico rispetto ai vecchi continenti capitalistici tradizionali. L'Asia è il secondo continente che presenta maggiori outliers. Oceania e America del Nord presentano alcuni anomalie, strano considerando il loro progresso industriale. America del Sud si presentano piuttosto omogenee, malgrado qualche eccezione.

```python
sns.boxplot( x='Continent', y='2016 [YR2016]', data=df, hue='Continent', legend=None)
plt.title('Percentuale della popolazione che ha accesso all\'elettricità per Continente nel 2016', fontsize=16)
plt.xlabel('Continente', fontsize=12)
plt.ylabel('2016 [YR2016]', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Istogramma con Matplotlib. Mostra la distribuzione della percentuale della popolazione con accesso all'elettricità nel 2016 tra i vari paesi. Sull'asse orizzontale (x) trovi gli intervalli di percentuali di accesso all'elettricità, mentre sull'asse verticale (y) trovi il numero di paesi che rientrano in quell'intervallo. Dall'istogramma possiamo osservare che un gran numero di paesi ha una percentuale molto alta di accesso all'elettricità, come indicato dalla barra più alta all'estrema destra. Ci sono anche paesi con un accesso all'elettricità molto basso, rappresentati dalle barre a sinistra. Questo evidenzia la disuguaglianza nell'accesso all'elettricità a livello globale e le sue conseguenze di sviluppo e benessere economico e sociale.

```python
plt.hist(df['2016 [YR2016]'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Distribuzione del consumo energetico nel 2016')
plt.xlabel('Consumo energetico')
plt.ylabel('Numero di Paesi')
plt.show()
```

Scatterplot o grafico a dispersione con Seaborn. Il grafico mostra la relazione tra il Continente e la percentuale di popolazione con accesso all'elettricità nel 2016. Ogni punto rappresenta un paese. Da questo grafico, si può notare come i paesi si distribuiscono per continente in termini di accesso all'elettricità, mostrando se ci sono raggruppamenti di paesi con accesso simile all'elettricità all'interno di un continente, o se c'è un'ampia variazione. Si evidenzia come in alcuni continenti i punti sono più sparsi (maggiore variabilità), mentre in altri sono più concentrati (minore variabilità). Ad esempio il continente africano si presenta come il continente con più divario in rapporto agli altri continenti, non solo all'esterno in rapporto agli contintenti ma anche al suo interno, infatti risultano esserci enormi disuguaglianze, rilevando come il continente africano non risulti omologato nella distribuzione di energia elettrica come ad esempio quello europeo, a parte qualche eccezione. L'Asia parzialmente presenta delle accomunanze con l'Africa. In un Continente ricco come l'america del nord stona la presenza di un outlier al 40%, probabilmente si riferisce alla zona centrale dell'america, sarebbe da approfondire. Lo stesso vale per l'Oceania con addirittuta un outlier di poco superiore al 20%.

```python
sns.scatterplot(data=df, y='2016 [YR2016]', x='Continent', hue='Continent', legend=None)
plt.title('Accesso all\'elettricità nel 2016 per Continente')
plt.xlabel('Continente')
plt.ylabel('Accesso all\'elettricità (% della popolazione)')
plt.xticks(rotation=45)
plt.show()
```

Stripplot o grafico a punti con Seaborn. Questo grafico mostra la distribuzione individuale della percentuale della popolazione con accesso all'elettricità nel 2016 per ciascun paese, raggruppati per continente. Ogni punto rappresenta un singolo paese. Dallo stripplot è possibile osservare la dispersione dei dati di accesso all'elettricità all'interno di ogni continente e di vedere la gamma di valori presenti e dove si concentrano la maggior parte dei paesi. È utile per identificare visivamente la variabilità all'interno dei gruppi (Continenti) e individuare eventuali outliers.

```python
sns.stripplot(x='Continent', y='2016 [YR2016]', data=df, jitter=True, hue='Continent', legend=None)
plt.title('Distribuzione individuale dell\'accesso all\'elettricità nel 2016 per Continente')
plt.ylabel('Accesso all\'elettricità (% della popolazione)')
plt.xlabel('Continente')
plt.xticks(rotation=45)
plt.show()
```

Barplot o grafico a barre con Seaborn. L'asse X (orizzontale) rappresenta i diversi continenti presenti nel dataset.L'asse Y (verticale) la percentuale media della popolazione con accesso all'elettricità per ciascun continente. Ogni barra corrisponde a un continente e l'altezza della barra indica il valore medio di accesso all'elettricità per tutti i paesi inclusi in quel continente nel 2016. La lunghezza della barra di errore dà un'idea della variabilità o incertezza intorno al valore medio mostrato dalla barra principale, aiutandoci a capire quanto è affidabile o rappresentativo il valore medio per ciascun continente. Guardando l'altezza delle barre, si può confrontare facilmente la percentuale media di accesso all'elettricità tra i vari continenti e identificare quali continenti hanno, in media, un accesso maggiore o minore all'elettricità nel 2016.

```python
sns.barplot(x='Continent', y='2016 [YR2016]', data=df, hue='Continent', legend=None)
plt.title('Accesso medio all\'elettricità nel 2016 per Continente')
plt.ylabel('Accesso medio all\'elettricità (% della popolazione)')
plt.xlabel('Continente')
plt.xticks(rotation=45)
plt.show()
```

Bar chart o grafico a barre con Matplotlib. Mostra la media della percentuale della popolazione con accesso all'elettricità nel 2016 per ogni continente. L'altezza di ogni barra rappresenta il valore medio di accesso all'elettricità per continente, permettendo in media un confronto diretto tra i continenti in termini di accesso medio all'elettricità nel 2016, evidenziando quali continenti hanno un accesso maggiore o minore accesso all'elettricità

```python
media_continente = df.groupby('Continent')['2016 [YR2016]'].mean()
plt.bar(media_continente.index, media_continente.values, color='orange')
plt.title('Accesso medio all\'elettricità nel 2016 per Continente')
plt.ylabel('Accesso medio all\'elettricità (% della popolazione)')
plt.xlabel('Continente')
plt.xticks(rotation=45)
plt.show()
```

Heatmap con Seaborn. Come già osservato con i barplot è evidente che alcuni continenti come l'Europa, che hanno un accesso medio molto più alto, l'elettricità risultati un bene scontato e di facile reperimento ai loro cittadini, rispetto ad altri come l'Africa. L'heatmap, a differenza dei barplot, ci offrono un'analisi descrittiva numerica più dettagliata.

```python
heatmap_data = df.groupby('Continent')['2016 [YR2016]'].mean().reset_index()
heatmap_data = heatmap_data.pivot_table(index='Continent', values='2016 [YR2016]')

sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".1f")
plt.title('Accesso medio all\'elettricità per Continente nel 2016')
plt.xlabel('Continente')
plt.ylabel('Anno (2016)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
